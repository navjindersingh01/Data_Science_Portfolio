```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import os
colors = sns.color_palette('pastel')
```

```python
graph_folder = "EDA_Graphs"

if not os.path.exists(graph_folder):
    os.makedirs(graph_folder)
else:
    pass
```

```python
def graph_save(name):
    save_path = os.path.join(graph_folder,name)
    plt.savefig(save_path,dpi = 300,bbox_inches = 'tight')
    
```

```python
df = pd.read_csv("salary_data_cleaned.csv")
```

```python
print(df.columns.tolist())
```

```python
cat_cols = ['seniority_level','status','industry','ownership']
cont_cols = ['avg_salary']
```

```python
sns.barplot(data = df, x = 'industry',y = 'avg_salary')
plt.title("Average Salary in Different industries",fontweight = 'bold',fontsize = 11)
plt.xticks(rotation = 90)
graph_save("Average_Salary_in_Different_industries")
plt.tight_layout()
```

```python
sns.histplot(data = df.avg_salary,kde = True)
plt.title("Normal Distribution of Salary",fontweight = 'bold',fontsize = 11)
graph_save("Normal_Distribution_of_Salary")
```

```python
sns.violinplot(data = df, x = 'seniority_level',y = 'avg_salary')
plt.title("Range and Concentration of Salary in Different levels",fontweight = 'bold',fontsize = 11)
graph_save("Range_and_Concentration_of_Salary_in_Different_leves")
```

```python
plt.pie(df.status.value_counts(), labels = df.status.unique().tolist(),autopct = '%1.2f%%',colors = colors)
plt.title("Job Distribution in Three different States of working",fontweight = 'bold',fontsize = 11)
plt.tight_layout
graph_save("Job_Distribution_in_Three_different_States_of_working")
```
